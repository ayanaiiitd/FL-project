## plot the global accuracy comparision (iid 2&4 epcohs)

```python
import matplotlib.pyplot as plt

def read_accuracies_from_file(filename):
    """
    Reads a text file and extracts accuracy values.
    
    Args:
        filename: Path to the text file.
        
    Returns:
        A list of accuracy values.
    """
    accuracies = []
    with open(filename, 'r') as file:
        for line in file:
            if "Round" in line and "%" in line:
                # Extract the accuracy value
                accuracy = float(line.split(":")[1].split('%')[0].strip())
                accuracies.append(accuracy)
    return accuracies


# File paths for the two text files
file1 = "textfiles/accuracyiid_all.txt"
file2 = "textfiles/accuracyiid_all1.txt"

# Read accuracies from each file
accuracies1 = read_accuracies_from_file(file1)
accuracies2 = read_accuracies_from_file(file2)

# Plot the accuracies
plt.figure(figsize=(10, 6))
rounds = range(1, len(accuracies1) + 1)

plt.plot(rounds, accuracies2, marker='s', label='Number of epochs = 4')

plt.plot(rounds, accuracies1, marker='o', label='Number of epochs = 2')

plt.xlabel('Rounds')
plt.ylabel('Accuracy (%)')
plt.title('Global Accuracy vs Rounds(iid)')
plt.legend()
plt.grid(True)
plt.savefig('textfiles/Accuracy_Comparison.png')
plt.show()
```
