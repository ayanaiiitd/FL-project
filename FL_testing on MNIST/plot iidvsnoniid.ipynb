## plot the global accuracy comparision (iid vs non-iid)

```python
import matplotlib.pyplot as plt

def read_accuracies_from_file(filename):
    """
    Reads a text file and extracts accuracy values.
    
    Args:
        filename: Path to the text file.
        
    Returns:
        A list of accuracy values.
    """
    accuracies = []
    with open(filename, 'r') as file:
        for line in file:
            if "Round" in line and "%" in line:
                # Extract the accuracy value
                accuracy = float(line.split(":")[1].split('%')[0].strip())
                accuracies.append(accuracy)
    return accuracies

def read_global_accuracies(filename):
    """
    Reads a text file and extracts global accuracy values.

    Args:
        filename: Path to the text file.

    Returns:
        A list of global accuracy values.
    """
    global_accuracies = []
    with open(filename, 'r') as file:
        for line in file:
            if "Global Accuracy" in line:
                # Extract the value after 'Global Accuracy:'
                accuracy = float(line.split("Global Accuracy:")[1].split('%')[0].strip())
                global_accuracies.append(accuracy)
    return global_accuracies

# File paths for the two text files
file1 = "textfiles/accuracyiid_all.txt"
file2 = "textfiles/accuracy_non_iid_22epochs.txt"

# Read accuracies from each file
accuracies1 = read_accuracies_from_file(file1)
accuracies2 = read_global_accuracies(file2)

# Plot the accuracies
plt.figure(figsize=(10, 6))
rounds = range(1, len(accuracies1) + 1)

plt.plot(rounds, accuracies1, marker='o', label='iid')
plt.plot(rounds, accuracies2, marker='s', label='non-iid')

plt.xlabel('Rounds')
plt.ylabel('Accuracy (%)')
plt.title('Global Accuracy vs Rounds(iid and non-iid comparision)')
plt.legend()
plt.grid(True)
plt.savefig('textfiles/Accuracy_Comparison.png')
plt.show()
```
