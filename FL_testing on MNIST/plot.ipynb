## plot the global accuracy comparision (local epoch mismatch)

```python
import matplotlib.pyplot as plt

def read_global_accuracies(filename):
    """
    Reads a text file and extracts global accuracy values.

    Args:
        filename: Path to the text file.

    Returns:
        A list of global accuracy values.
    """
    global_accuracies = []
    with open(filename, 'r') as file:
        for line in file:
            if "Global Accuracy" in line:
                # Extract the value after 'Global Accuracy:'
                accuracy = float(line.split("Global Accuracy:")[1].split('%')[0].strip())
                global_accuracies.append(accuracy)
    return global_accuracies

# File paths for the three text files
file1 = "textfiles/accuracy_3clients_2mismatch.txt"
file2 = "textfiles/accuracy_3clients_singlemismatch.txt"
file3 = "textfiles/accuracy_3clients.txt"

# Read global accuracies from each file
global_accuracies1 = read_global_accuracies(file1)
global_accuracies2 = read_global_accuracies(file2)
global_accuracies3 = read_global_accuracies(file3)

# Plot the global accuracies
plt.figure(figsize=(10, 6))
rounds = range(1, len(global_accuracies1) + 1)

plt.plot(rounds, global_accuracies3, marker='^', label='No mismatch of epocs among clients')
plt.plot(rounds, global_accuracies2, marker='s', label='1 clients\' epoch mismatch')
plt.plot(rounds, global_accuracies1, marker='o', label='2 clients\' epoch mismatch')

plt.xlabel('Global Rounds')
plt.ylabel('Global Test Accuracy (%)')
plt.title('Global Test Accuracy Comparison Over Federated Learning Rounds(for 3 Clients)')
plt.legend()
plt.grid(True)
plt.savefig('textfiles/Global_Accuracy_Comparison.png')
plt.show()
```
